```python
#checking numbers of players from each nationality
fifa['nationality'].value_counts()
```

```python
#checking top 10 countries with most players
fifa['nationality'].value_counts()[0:10]
```

```python
#creating a bar plot of top 5 countries with most players

#creating lists
list(fifa['nationality'].value_counts()[0:5].keys())

#creating the bar plot
plt.figure(figsize=(5,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()), list(fifa['nationality'].value_counts()[0:5]), color='green')
plt.title('countries vs players')
plt.xlabel('countries')
plt.ylabel('number of players')
plt.show()
```

```python
#let us know what is the salary of top 5 player
player_salary = fifa[['short_name', 'wage_eur']]
player_salary.head()
```

```python
#let us know what is the salary of last 5 players
player_salary.tail()
```

```python
#let us sort players according to their wages in descending order
player_salary = player_salary.sort_values(by=['wage_eur'], ascending=False)
player_salary.head()
```

```python
#check which players gets least salary
player_salary.tail()
```

```python
#let us create a bar plot for players according to their salary

#let plot the bar graph
plt.figure(figsize=(5,5))
plt.bar(list(player_salary['short_name'].value_counts()[0:5].keys()), list(player_salary['wage_eur'].value_counts()[0:5]))
plt.title('salary vs player')
plt.xlabel('player name')
plt.ylabel('salary in euro')
plt.show()
```
