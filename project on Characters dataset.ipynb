```python
#loading the required libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
#loading the .csv file to work upon
marvel = pd.read_csv(r"C:\Users\spars\Desktop\charcters_stats.csv")
```

```python
#checking the first 5 rows
marvel.head()
```

```python
#checking the total number of rows and columns
marvel.shape
```

```python
#checking all the good superheroes
good_supes = marvel[marvel['Alignment']=='good']
good_supes.head()
```

```python
#sorting good superheores according to their speed
good_supes.sort_values(by=['Speed'], ascending=False)
```

```python
#sorting good superheores according to their total score
greatest_supes = good_supes.sort_values(by=['Total'], ascending=False)
greatest_supes.head()
```

```python
#plotting a bar graph for top 5 superheroes with highest total score

#creating a list
list(greatest_supes['Name'])[0:5]

#plotting the graph
plt.figure(figsize=(10,5))
plt.bar(list(greatest_supes['Name'])[0:5], list(greatest_supes['Total'])[0:5], color=['green', 'red', 'yellow', 'black', 'blue'])
plt.title('supes vs score')
plt.xlabel('superheroes')
plt.ylabel('total score')
plt.show()
```

```python
#checking all the villains
villains = marvel[marvel['Alignment']=='bad']
villains.head()
```

```python
#sorting villains according to their total score
greatest_villains = villains.sort_values(by=['Total'], ascending=False)
greatest_villains.head()
```

```python
#making a histogram for top 5 villains with highest total score
plt.hist(greatest_villains['Total'][0:5], bins=30)
plt.title('villains')
plt.xlabel('total score')
plt.show()
```
